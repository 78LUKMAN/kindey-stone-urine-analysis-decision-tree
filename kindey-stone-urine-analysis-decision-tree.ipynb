```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
dataset = 'dataset/data-iris.csv'
```

```python
df = pd.read_csv(dataset, sep=';')
df.head()
```

```python
feature = df.iloc[:, 1:-1]
target = df.iloc[:, -1]
# print(target)
print(feature)
```

```python
decisiontree = DecisionTreeClassifier(random_state=0, max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0, max_leaf_nodes=None, min_impurity_decrease=0)
```

```python
model = decisiontree.fit(feature,target)
```

```python
observation = [[5, 4, 3, 2]]  # Contoh data observasi yang akan diprediksi
prediction = model.predict(observation)
prediction_proba = model.predict_proba(observation)
```

```python
print("Prediksi:", prediction)
print("Probabilitas prediksi:", prediction_proba)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
plt.figure(figsize=(20, 10))  
plot_tree(model, 
          feature_names=feature.columns,
          class_names=target.unique(),
          filled=True,              
          rounded=True,                 
          fontsize=12)         
plt.savefig('decision-tree.png', format='png')
plt.show()
```
